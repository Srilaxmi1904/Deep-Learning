```python
import cv2
import os
import numpy as np
import pickle
import face_recognition
```

```python
cam=cv2.VideoCapture(-1)
cam.set(3,198)
cam.set(4,146)
```

```python
imgbg=cv2.imread('/home/srilaxmi/Documents/DEEP LEARNING/open_cv/PROJECT_1/attendance.jpeg')
```

```python
#importing mode pictures as a list:
modepath1='/home/srilaxmi/Documents/DEEP LEARNING/open_cv/PROJECT_1/modes'
modepath=os.listdir(modepath1)
modelist=[]

print(modepath)

for i in modepath:
    modelist.append(cv2.imread(os.path.join(modepath1,i)))

print(len(modelist))
#will add this to the backgroud image
```

```python
#imported the enoded file:
file=open('encodedfile','rb')
knownfacesplusid=pickle.load(file)
file.close()

#seperating "knownfaces" and "id":
knownfaces,id=knownfacesplusid
print(id)
# print(knownfaces)
```

```python
facecas=cv2.CascadeClassifier('/home/srilaxmi/Documents/DEEP LEARNING/open_cv/Face Detection/haarcascade_frontalface_default.xml')
```

```python
while True:

    success,frame=cam.read()

    small=cv2.resize(frame,(0,0),None,0.50,0.50)  #making the image a bit smaller so that computation time is reduced.
    small=cv2.cvtColor(small,cv2.COLOR_BGR2RGB)

    currentface=face_recognition.face_locations(small)  #getting the position/location of the images
    encocurframe=face_recognition.face_encodings(small,currentface)

    imgbg[64:64+144,24:24+176]=frame
    imgbg[21:21+205,266:266+143]=modelist[0]

    face=facecas.detectMultiScale(imgbg,scaleFactor=1.1,minNeighbors=4) 


    for facenco,faceloc in zip(encocurframe,currentface):
        match=face_recognition.compare_faces(knownfaces,facenco)
        distance=face_recognition.face_distance(knownfaces,facenco)   #cause lesser the distance, means more accuracy
        print('match',match)
        print('distance',distance)

        index=np.argmin(distance)
        print(index) 
        
        for (x,y,w,h) in face:
            cv2.rectangle(imgbg,(x,y),(x+w,y+h),color=(0,255,0),thickness=2)

    # cv2.imshow('Face',frame)
    cv2.imshow('img',imgbg)
    cv2.waitKey(1)
    # cv2.destroyAllWindows()
```
